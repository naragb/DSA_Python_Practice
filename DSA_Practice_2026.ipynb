```python
# Duplicate within K Distance in an Array

# Naive Approach 
def check_duplicates_within_k(arr, k):
    n = len(arr)
    
    for i  in range(n):
        for c in range(1, k + 1):
            j = i + c
            
            if j<n and arr[i] == arr[j]:
                return True
            
    return False

arr = [10, 5, 3, 4,3,5,6]
print("Yes" if check_duplicates_within_k(arr, 3) else "No")
```

```python
# Expected approach using hash set
def check_duplicates_within_k(arr, n, k):
    myset= set()
    
    for i in range(n):
        
        if arr[i] in myset:
            return True
        
        myset.add(arr[i])
        
        if(i >= k):
            myset.remove(arr[i - k])
    return False 

if __name__ == "__main__":
    arr=[10, 5, 3, 4, 3, 5, 6]
    n=len(arr)
    if (check_duplicates_within_k(arr, n,3)):
        print("Yes")
    else:
        print("No")
```

```python
# Rearrange array such that even positioned are greater than odd

# Approach 1: Assign Maximum Elements to Even positions 

def rearrangeArray(arr):
    arr.sort()
    
    n = len(arr)
    result =[0] * n
    ptr1, ptr2 =0, n-1
    
    for i in range(n):
        if (i + 1) % 2 ==0:
            result[i] = arr[ptr2]
            ptr2 = -1
            
        else:
            result[i] = arr[ptr1]
            ptr1 += 1
            
    return result

arr = [1, 2, 2, 1]
res = rearrangeArray(arr)
print(res)
```

```python
# Approach 2

def rearrangeArray(arr):
    n = len(arr)
    
    for i in range(1, n):
        if (i + 1) % 2 == 0:
            if arr[i] < arr[i - 1]:
                arr[i], arr[i - 1] = arr[i - 1], arr[i]
        else:
            if arr[i] > arr[i - 1]:
                arr[i], arr[i - 1] = arr[i - 1], arr[i]
                
    return arr

if __name__ =="__main__":
    inputArray = [1,2,2,1]
    resultArray = rearrangeArray(inputArray)
    print(" ".join(map(str, resultArray)))
```

```python
# Challenge 12: array
def neighbor_avg3(values):
        
    if len(values) < 3:
        return values[:]
    
    out = []
    
    for i in range(len(values)):
        
        if i == 0 or i == len(values)-1:
            out.append(values[i])
            
        else:
            res = (values[i-1] + values[i] + values[i+1])/3
            out.append(res)
    
    return out 
```

```python
# Practice: Sum of All Subarrays

# Method 1:

def subarraySum(arr):
    n = len(arr)
    result = 0
    
    for i in range(n):
        temp = 0
        
        for j in range(i, n):
            temp += arr[j]
            result += temp
            
    return result

if __name__="__main__":
    arr = [1,4,5,3,2]
    print(subarraySum(arr))
```

```python
# Method 2

def subarraySum(arr):
    n = len(arr)
    result = 0
    
    for i in range(n):
        result += arr[i] * (i+1) *(n-i)
        
    return result

if __name__="__main__":
    arr = [1,4,5,3,2]
    print(subarraySum(arr))
```

```python
# Date: Jan 17, 2026
# G4G Practice: Unique Number I

def findUnique(arr):
    res = 0
    
    for num in arr:
        res ^= num
    return res

if __name__ =='__main__':
    arr = [2,3,4,5,3,4,5]
    print(findUnique(arr))
```

```python
# Date: Jan 17, 2026
# G4G Practice: Find the Missing Number
# Method 1 

def missingNum(arr):
    n = len(arr) + 1
    xor1 = 0
    xor2 = 0
    
    for i in range(n - 1):
        xor2 ^= arr[i]
        
    for i in range(1, n+1):
        xor1 ^= i
        
    return xor1 ^ xor2

if __name__ == '__main__':
    arr = [8, 2, 4, 5, 3, 7, 1]
    res = missingNum(arr)
    print(res)
```

```python
def findTwoElement(arr):
    n = len(arr)
    xorVal = 0  

    # get the xor of all array elements
    # And numbers from 1 to n
    for i in range(n):
        xorVal ^= arr[i]
        xorVal ^= (i + 1)  # 1 to n numbers

    # get the rightmost set bit in xorVal
    setBitIndex = xorVal & ~(xorVal - 1)
    
    x, y = 0, 0

    # now divide elements into two sets 
    # by comparing rightmost set bit
    for i in range(n):
      
        # decide whether arr[i] is in first set 
        # or second
        if arr[i] & setBitIndex: 
            x ^= arr[i] 
        else: 
            y ^= arr[i]
      
        # decide whether (i+1) is in first set 
        # or second
        if (i + 1) & setBitIndex: 
            x ^= (i + 1) 
        else: 
            y ^= (i + 1)

    # x and y are the repeating and missing values.
    # to know which one is what, traverse the array 
    xCnt = sum(1 for num in arr if num == x)
    
    if xCnt == 0:
        missing, repeating = x, y
    else:
        missing, repeating = y, x

    return [repeating, missing]

if __name__ == "__main__":
    arr = [3, 1, 3]
    ans = findTwoElement(arr)

    print(ans[0], ans[1])
```

```python
# Date: Jan 17, 2026
# G4G Practice: Find the Missing Number
# Method 2 

def missingNum(arr):
    n = len(arr) + 1
    
    totalSum = sum(arr)
    
    expSum = n *(n+1)//2
    
    return expSum - totalSum


if __name__ == '__main__':
    arr = [8, 2, 4, 5, 3, 7, 1]
    res = missingNum(arr)
    print(res)
```

```python
# Date: Jan 17, 2026
# G4G Practice: Find the Missing Number
# Method 3


def missingNum(arr):
    n = len(arr) + 1
    
    hash = [0] * (n + 1)
    
    for i in range(n - 1):
        hash[arr[i]] += 1
        
    for i in range(1, n + 1):
        if hash[i] == 0:
            return i
        
    return -1
    

if __name__ == '__main__':
    arr = [8, 2, 4, 5, 3, 7, 1]
    res = missingNum(arr)
    print(res)
```

```python
# Date: Jan 17, 2026
# G4G Practice: Find the Missing Number
# Method 4


def missingNum(arr):
    n = len(arr) + 1
    
    for i in range(1, n+1):
        found = False
        for j in range(n-1):
            if arr[j] == i:
                found = True
                break
            
        if not found:
            return i
        
    return -1
    

if __name__ == '__main__':
    arr = [8, 2, 4, 5, 3, 7, 1]
    res = missingNum(arr)
    print(res)
```

```python
# Date: Jan 17, 2026
# G4G Practice: Only Repeating From 1 To n-1
# Method 1

def findDuplicate(arr):
    n = len(arr)
    
    for i in range(n):
        for j in range(i + 1, n):
            if arr[i] == arr[j]:
                return arr[i]
    return -1

if __name__ == '__main__':
    arr = [1,3,2,3,4]
    print(findDuplicate(arr)) 
```

```python
# Date: Jan 17, 2026
# G4G Practice: Only Repeating From 1 To n-1
# Method 2

def findDuplicate(arr):
    
    arr.sort()
    
    for i in range(len(arr) - 1):
        if arr[i] == arr[i + 1]:
            return arr[i]
        
    return -1 

if __name__ == "__main__":
    arr = [1, 3, 2, 3, 4]
    print(findDuplicate(arr))
            
```

```python
# Date: Jan 17, 2026
# G4G Practice: Maximum Subarray Sum - Kadane's Algorithm
# Method 1

def maxSubarraySum(arr):
    res = arr[0]
    
    maxEnding = arr[0]
    
    for i in range(1, len(arr)):
        maxEnding = max(maxEnding + arr[i], arr[i])
        
        res = max(res, maxEnding)
        
    return res

if __name__ == "__main__":
    arr = [2, 3, -8, 7, -1, 2, 3]
    print(maxSubarraySum(arr))
    
```

```python
def count_users(events: list[int]) -> dict[int, int]:
    counts = {}
    
    for num in events:
        if num in counts: 
            counts[num] = counts.get(num, 0) + 1
        else:
            counts[num] = 1
    return counts
           
```
